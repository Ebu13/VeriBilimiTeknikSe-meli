```python
from gensim.scripts.glove2word2vec import glove2word2vec
from gensim.models import KeyedVectors
from google.colab import files
```

```python
glove_input = files.upload()
glove_input = list(glove_input.keys())[0]
word2vec_output = 'glove.6B.100d.word2vec'
glove2word2vec(glove_input, word2vec_output)
```

```python
model = KeyedVectors.load_word2vec_format(word2vec_output, binary = False)
```

```python
model['of']
```

```python
model.most_similar('of')
```

```python
model.most_similar(positive = ['the', 'this'], negative=['to'], topn = 1) # woman - man + son = daugther; king - man + woman = queen
```
